```python
label_cols = ['brand', 'transmission', 'ext_col', 'int_col']
label_encoders = {col: LabelEncoder() for col in label_cols}
for col in label_cols:
    df[col + '_label'] = label_encoders[col].fit_transform(df[col])
    classes_seen = set(label_encoders[col].classes_)

    def handle_unseen_labels(value):
        if value in classes_seen:
            return label_encoders[col].transform([value])[0]
        else:
            return -1  # Special category for unseen labels
    
    test_df[col + '_label'] = test_df[col].apply(handle_unseen_labels)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(128 ,activation='relu' ,input_shape=(X_train.shape[1] ,)),
    tf.keras.layers.Dropout(0.2),  # Dropout layer for regularization

    tf.keras.layers.Dense(64 ,activation='relu'),
    tf.keras.layers.Dropout(0.2),  # Dropout layer for regularization

    tf.keras.layers.Dense(32 ,activation='relu'),
    tf.keras.layers.Dropout(0.2),  # Dropout layer for regularization
    
    tf.keras.layers.Dense(16 ,activation='relu'),
    tf.keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
```

```python
import keras_tuner as kt
from keras import Sequential
from keras.layers import Dense ,Dropout
```

```python
def build_model(hp):
    model = Sequential()
    
    counter = 0
    
    for i in range(hp.Int('num_layers' ,min_value=1 ,max_value=10)):
        if counter == 0:
            model.add(Dense(units=hp.Int('units' + str(i), min_value=32, max_value=512, step=32),
                                            activation= hp.Choice('activation' + str(i) ,values=['tanh','relu','sigmoid']), 
                                            input_shape=(X_train.shape[1],)))
            
            model.add(Dropout(hp.Float('dropout' + str(i), min_value=0.0, max_value=0.9, step=0.1)))
            
        else:
            model.add(Dense(units=hp.Int('units' + str(i), min_value=32, max_value=512, step=32),
                                            activation= hp.Choice('activation' + str(i) ,values=['tanh','relu','sigmoid'])))
            
            model.add(Dropout(hp.Float('dropout' + str(i), min_value=0.0, max_value=0.9, step=0.1)))
                      
    counter = counter + 1
            
    model.add(Dense(1))
                      
    model.compile(optimizer= hp.Choice('optimizer' ,values=['adam','sgd','nadam','adadelta','rmsprop']), 
                  loss='mean_squared_error', 
                  metrics=['mae'])
    return model
```

```python
tuner = kt.GridSearch(build_model ,
                        objective='val_loss' ,
                        max_trials=50 ,
                        directory='mydir',
                        project_name='final')

stop_early = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)

tuner.search(X_train, y_train, epochs=50, validation_data=(X_test ,y_test), callbacks=[stop_early])
```

```python
best_hps = tuner.get_best_hyperparameters()[0].values
best_hps
```

```python
model = tuner.get_best_models(num_models=1)[0]
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=500, validation_data=(X_test ,y_test), callbacks=[stop_early] ,initial_epoch=100 ,batch_size=32)
```

```python
y_pred = model.predict(X_test)
y_pred = y_pred.flatten()

r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print('R2 Score:', r2)
print('MAE:', mae)
```
